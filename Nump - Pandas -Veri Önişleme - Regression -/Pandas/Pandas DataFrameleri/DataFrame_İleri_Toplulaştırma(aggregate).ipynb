```python
import pandas as pd 
import numpy as np
```

```python
df = pd.DataFrame({"degisken1" : [3 ,2 ,5 ,2 ,6, 7], 
                  "degisken2" : [5 ,24 ,2 ,65, 45 ,5],
                  "gruplar" : ["index1" ,"index2" ,"index3" ,"index1" ,"index2" ,"index3"]})
```

```python
df
```

```python
df.groupby("gruplar").sum()
```

```python
df.groupby("gruplar").aggregate(["min" ,np.median , "max"]) 
#pandasta olmuyan işlemlerin başına ktüphane isminin gelmesine gerek yoktur
```

```python
df.groupby("gruplar").aggregate({"degisken1" : ["min" ,"max"] ,
                                "degisken2" : ["count" ,"mean"]})
```
