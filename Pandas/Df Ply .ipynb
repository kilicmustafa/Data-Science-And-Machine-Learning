```python
from dfply import * 
```

```python
df = diamonds 
df >> head(6)
# df in içinden ilk 6 değeri getir
```

```python
df >> select(X.cut , X.color) >> head(5)
# df in içinden Cut ve color sutunlarının ilk 5 değerini getir
```

```python
df >> select(1 ,2 ,3) >> head(5)
# dataframenin içinden 1 , 2 ,3 indexli sutunların ilk 5 değerini getir
```

```python
df >> select(["x" ,"y" ,"z" ,"color" ,"cut"])
# dataframenin içinden listedeği sutunları getirir
```

```python
df >> select(starts_with("c")) >> head(5)
# dataframenin içinden sutununun ismi " c " ile başlıyanların ilk 5 satırını getirir
```

```python
df >> select(ends_with("e")) >> tail(5)
# dataframenin içinden sutununun ismi e ile biten sutunların son 5 satırının(değerini) getirir
```

```python
df >> select(contains("a")) >> tail(4)
# dataframenin içinde sutununun adında a harfi bulunanların sondan 4 satırını ( değerini ) getirir
```

```python
df >> select(columns_between("x" ,"z")) >> head(5) 
# dataframenin içinden sutunlarından x ile z arasında olan sutunların ilk 5 değerini getirir
```

```python
df >> select(~ X.carat , ~X.color ) >> head(5)
# dataframe içerisinde carat ve color sutunları haricindeki tüm sutunları getirir
```

```python
df >> row_slice([10 ,15])
# dataframenin içinden 10 ve 15. satırları getirir
```

```python
df >> group_by( X.cut) >> row_slice(5)
# dataframenin içinden Cut değişkenie göre grupla ve içinden grupları farklı 5 er cut değeri getir 
```

```python
df >> group_by(X.cut , X.color) >> row_slice(5)
# dataframeyi  cut ve colora göre grupla ve cut ve color değerleri birbirinden farklı her group eşteniği için 5 değer geitr
```

```python
df >> sample(n = 4 , replace = False)
# df içerisinde rasgele 4 tane değer getir ve bu değerler birbiriyle asla aynı olmasın 
```

```python
df >> sample( n = 5 , replace = True) 
# DataFrame içinden 5 tane rasgele değer getir ve bu değerler tekrarlanarak aynı değerlerden birden cok olablicek sekilde getirilmesine izin ver
```

```python
df >> distinct(X.color) 
#  dataframe içerisinde color rengi tekrarsız olacak şekilde veri getirir
```

```python
df >> mask(X.cut == "Ideal" ) 
# dataframenin içinden Cut değeri Ideal olanları getirir
```

```python
df >> mask(X.cut  == "Ideal" ,X.color == "E" , X.table < 55)
# df içinden ustteki sartları saglayanları getirir 
```

```python
df >> filter_by(X.cut  == "Ideal" ,X.color == "E" , X.table < 55)
# mask fonksiyonu ile aynı işlemi görür
```

```python
df >> mutate(x_plus = (X.x + X.y))  >> head(5)
# df içerisine x_plus diye yeni değişken ekleyip bunun değerlerine X ve Y nin toplamlarını atar
```

```python
df >> mutate(y_plus = (X.y * X.z)) >> head(5)
# dataframeye y_plus adında yeni bir sutun ekleyip bunun değerlerini y ile z nin çarpımından üretir
```

```python
df >> mutate(z_plus = (X.z - X.z)) >> select(columns_from("x")) >> head(5)
# df içine z_plus diye değer ekler ve " x ' den sonraki ' " sutunları getirir
```

```python

df >> transmute(y_plus = (X.y * X.z),z_plus = (X.z - X.z)) >> head()
# df içinden y_plus ve z_plus değerleri oluşturulur ve " transmute  " sasyesinde sadece oluşturulan sutunlar getirilir
```

```python
df >> group_by(X.cut) >> arrange(X.price) >> head(2)
#dataframenin içnden cut'a göre gruplanmış değerleri " arrange  " fonksiyonu saysinde sıralayarak getirir
#her grup değerinden 2 şer olacak şekilde
```

```python
df >> group_by("cut") >> arrange("price") >> ungroup() >> mask(X.carat < 0.23 )
# df'i cut'a göre gurpla ve priceye göre sırala ardından gurubu bozup filtreleme işlemine tabi tut 
```

```python
df >> summarize(price_ortalama = X.price.mean() , price_ss = X.price.std() )
# df içindeki pricenin ortalamasını ve sandart sapmasını heapla ve bunlar için bir sutun oluştur ardından ekranda gösteR
```

```python
df >> group_by("cut" ) >> summarize ( p_ort = X.price.mean() , p_ss = X.price.std())
# üstteki islemin grupby yapılmış hali
```
