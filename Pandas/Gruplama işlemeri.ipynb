## aggregate ,filter ,transform ve apply işlemleri

```python
df = pd.DataFrame(
    
    {
        "gruplar" : ["a" ,"b" ,"a" ,"c" ,"a" ,"b"],
        "deg_1" : [1 ,2 ,3 ,5 ,7 ,9],
        "deg_2" : [10 ,20 ,30 ,40 ,50 ,60]
      
       
    }
)
df
```

```python
df.groupby("gruplar").sum()
```

Aggregate

```python
df.groupby("gruplar").aggregate(["min" ,"max"])
# aggregate fonksiyonu sayesinde her bir değişken için özel fonksiyon hesaplamaları yaptırdık
```

```python
df.groupby("gruplar").aggregate({"deg_1" : ["min" , np.median],
                                "deg_2" : ["max" ,np.mean]})
# aggregate sözlük kullanımı sayesinde her bir değişken için kiselleştirilmiş fonksyon hesaplattırdık
```

Filter

```python
def filt_fonk(fonk):
    return fonk["deg_1"].mean() > 4

df.groupby("gruplar").filter(filt_fonk)

# burada bir flitreleme işlemi yaptık 
```

Transform

```python
df.groupby("gruplar").transform(lambda x : x - x.mean())
# transform işlemi ile tüm hücrelere istediğimiz işlemi uyguluyabiliriz
```

```python
df.groupby("gruplar").transform(lambda x : x == 0)
# sıfır olan değerleri kontrol ediyoruz
```

```python
df.groupby("gruplar").transform(lambda x : x - x)
# tum değerleri sıfırlıyoruz
```

Apply fonksiyonu

```python
df.groupby("gruplar").apply(filt_fonk)
```

```python
df.groupby("gruplar").apply(np.mean)
# apply fonksiyonu ile gruplara ortalama gibi işlemleri yaptırailiriz
```
