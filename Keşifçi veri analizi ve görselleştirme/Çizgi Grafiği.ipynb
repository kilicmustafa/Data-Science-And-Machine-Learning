```python
df.timepoint.describe()
# sadece standart sapma değeri ondalıklı demekki bu gruplayıcı birşey
```

```python
df.groupby("timepoint")["signal"].count()
# her bir timepointte kaç tane sinyal var
```

```python
sns.lmplot(data = df ,x = "timepoint" , y = "signal" ,hue = "subject");
# mantıksız ama güzel grafik 
```

```python
sns.pairplot(df , hue = "subject" );
```

```python
sns.scatterplot(data = df ,x = "signal" , y= "timepoint"  ,hue = "subject")
# saçma oldu bu 
```

```python
# konuya dönmelim merak ettiklerimize baktığımıza göre 


```

```python
sns.lineplot(data = df ,x = "timepoint" , y = "signal");
# timepoint göre signal in değişimini gördük anacak bu bi 3. boyutu ekliyelim bakalım ne olacak
```

```python
sns.lineplot(data = df , x = "timepoint" ,y = "signal" ,hue = "event" );
# evet 3. boyutu eklediğimizde burada event değişkenindek stim kategorısi signalin ortalamasını yükseltiyor 
```

```python
sns.lineplot(data = df , x = "timepoint" , y = "signal" , hue = "event" , style = "event");
# evet burda evet de bununan değişkenlere özel şekiller verdik
```

```python
sns.lineplot(data = df  ,x = "timepoint" , y = "signal" , hue = "region" , style = "event");
# evet bu grafige bakarak event içerisindeki cue ler birbirlerine yakınken , stim ogreleri birbirinden daha uzakta seyrediyor 
```

```python
sns.lineplot(data = df , x = "timepoint" ,y = "signal" ,hue = "event" 
            ,style = "event"
            ,markers = True 
            );
```

```python
sns.lineplot(data = df , x = "timepoint" ,y = "signal" ,hue = "event" 
            ,style = "event"
            ,markers = True ,dashes = False
            );

# evet boylede ortalamalarını noktalamış olduk
```
