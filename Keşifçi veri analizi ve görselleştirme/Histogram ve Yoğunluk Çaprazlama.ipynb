```python
import pandas as pd 
import seaborn as sns
diamonds = sns.load_dataset("diamonds")
df = diamonds.copy()
df.head(5)
```

```python
df.describe().T
```

```python
df.price.index
```

```python
sns.distplot(df.price ,kde = False);
```

```python
sns.kdeplot(df.price , shade = True) 
# shade = bolgeyi tarama yani renklendirme
```

```python
# yukarıdaki değişkenin altında dönen olayları bilmediğimiz için yorumda bulunursak
#yanılmış olabiliriz bunun için grafiğe 3. boyutu katarak yanılgı payımızı azaltabiliriz
```

```python
(sns
 .FacetGrid(df,
               hue = "cut" , # boyut olarak ekliyeceğimiz değişken
               height = 5 ,
               xlim = (0 ,10000) # x ekseninin 0 ile 1000 arasına yoğunlaşacağız cunku diğer yerlerde anorlmallik yok
           )
 .map(sns.kdeplot , "price" ,shade = True)
 .add_legend() # bu fonksiyon ile cut değişkeninin isimlerini ekliyeceğiz 
 

);


# yukarıda FacetGrid Fonksiyonunu boyut eklemek için kullanıyoruz ,map fonksiyonu ilede bunları eşliyoruz
```

```python
sns.catplot(data = df , x = "cut" , y = "price" , hue = "color" ,kind = "point");
# burada cut ve color değişkenlerini gruplatıp price değişkenine denk gelen ortalamalarını aldık 
# kint ile grafiğin cizim türünün noktalama olacağını belirttik
```

```python
(sns
.FacetGrid(df,
row = "color",
          hue = "cut",
          height = 5,
          xlim = (0,10000)
    )

.map(sns.kdeplot ,"price" ,shade = True)
.add_legend()
)
```
