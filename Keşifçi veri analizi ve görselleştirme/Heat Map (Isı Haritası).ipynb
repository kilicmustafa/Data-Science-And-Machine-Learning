```python
import seaborn as sns
import pandas as pd 
flights = sns.load_dataset("flights")
df = flights.copy()
df.tail(5)
```

```python
df.info()
```

```python
pivot = df.pivot(index = "month" , columns = "year" , values = "passengers")
pivot
```

```python
# eger pivot table formatına sokmaz isk ısı haritası çalışmaz

```

```python
sns.heatmap(pivot);
```

```python
sns.heatmap(pivot ,annot = True ,fmt = "d" );
# annot sayı değerlerini getirmek için fmt ise sayı değerlerini düzgün getirmel içindir 
# yani decimal olarak 
```

```python
sns.heatmap(pivot , annot = True , fmt = "d") ;
```

```python
sns.heatmap(pivot , annot = True , fmt = "d" , cbar = False) ;
# c barı kaldırır
```
