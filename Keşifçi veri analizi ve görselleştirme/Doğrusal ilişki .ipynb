```python
import pandas as pd 
import numpy as np
import seaborn as sns 

tips = sns.load_dataset("tips")
df = tips.copy()

df.head(5)
```

```python
import matplotlib.pyplot as plt
```

```python
sns.lmplot(data = df ,x = "total_bill" ,y = "tip");
# hesap ile bahşiş arasındaki doğrusal iliştiyi eğim ve sapmalar ile gösterir
# grafikteki açık mavilikler sapma oranlarıdır
```

```python
sns.lmplot(data = df , x = "total_bill" , y = "tip" , hue= "smoker");
#sigara içip içmeme durumuna gore bahşiş ve hesap odeme ücreti arasındaki ilişkiyi ifade eden dorular
```

```python
sns.lmplot(data = df , x = "total_bill" , y = "tip" , hue = "smoker" , col = "time");
# sigara içme durumu ve zaman kırılımları ile ödenen hesap ve bahşiş oranının doğrusunu ele almış olduk 
```

```python
sns.lmplot(data = df , x = "total_bill" , y = "tip" , hue = "smoker" ,col = "time" ,row = "sex" );
#üstteki grafiğe cinsiyet kırılımınıda kattık 
```

```python
sns.lmplot(data = df ,x = "total_bill" , y = "tip" , hue = "day" , col = "time");
# gün ve zamana göre hesap ödeme grafiği 
```
