```python
import seaborn as sns
iris = sns.load_dataset("iris")
df = iris.copy()
df.tail(5)
```

```python
df.info()
```

```python
df.species.value_counts()
```

```python
df.species.value_counts().plot.barh()
```

```python
sns.pairplot(data =df );
```

```python
sns.pairplot(data = df , hue = "species");
# species kırılımına göre scatterplot görselleri
```

```python
sns.pairplot(data = df , hue = "species" ,markers = ["o" ,"s" ,"d"]);
#burda markers değişkeni ile grafiklerimizin üzerinde bulunan simgelere şekil verdik
```

```python
sns.pairplot(data = df , hue = "species" ,kind = "reg" );
#kind ile doğrusal eğrileri aldık 
```
