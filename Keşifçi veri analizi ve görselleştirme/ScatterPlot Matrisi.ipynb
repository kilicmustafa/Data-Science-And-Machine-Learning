```python
import seaborn as sns
iris = sns.load_dataset("iris")
df = iris.copy()
df.tail(5)
```

```python
df.info()
```

```python
df.species.value_counts()
```

```python
df.species.value_counts().plot.barh()
```

```python
sns.pairplot(data =df );
```

```python
sns.pairplot(data = df , hue = "species");
# species krlmna gre scatterplot grselleri
```

```python
sns.pairplot(data = df , hue = "species" ,markers = ["o" ,"s" ,"d"]);
#burda markers deikeni ile grafiklerimizin zerinde bulunan simgelere ekil verdik
```

```python
sns.pairplot(data = df , hue = "species" ,kind = "reg" );
#kind ile dorusal erileri aldk 
```
