```python
import pandas as pd 
import seaborn as sns 
tips = sns.load_dataset("tips")
df = tips.copy()
df.head()
```

```python
# Yoğunluk grafiği ile kutu grafiğinin kesişimi olarak düşünülebilir
```

```python
# dikey olarak bakınca box plot , yatay olarak bakınca yoğunluk grafiği 
```

```python
sns.catplot(data = df ,y = "total_bill"  , kind= "violin")
# toplam kazanç grafiği
```

```python
sns.catplot(data = df , x = "day" , y = "total_bill" ,kind = "violin")
```

```python
sns.catplot(data = df , x = "day" , y = "total_bill" , hue = "smoker" ,kind = "violin")
```
