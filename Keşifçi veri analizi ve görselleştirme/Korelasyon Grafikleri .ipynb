## ScatterPlot Kullanımı 

```python
import pandas as pd 
import numpy as np
import seaborn as sns 

tips = sns.load_dataset("tips")
df = tips.copy()

df.head(5)
```

```python
sns.scatterplot(data = df , x = "total_bill" ,y ="tip" );
# hesap odeme ücretinin bahşişlere olan etkisi
```

```python
sns.scatterplot(data = df  ,x = "total_bill" ,  y = "tip" , hue = "day");
# günlere göre hesap ,bahşiş grafiği 
```

```python
sns.scatterplot(data = df , x = "total_bill" , y = "tip" , hue ="day" ,style = "day") ;
# günlere göre hesap ,bahşiş grafiği stilli halli
```

```python
sns.scatterplot(data = df , x = "total_bill" ,y = "tip" , hue = "day" , style = "sex" );
# günlere ve cinsiyet kırılımlarına göre bahşiş ve hesap ilişkisi
```

```python
sns.scatterplot(data = df , x = "total_bill" ,y = "tip" ,  hue = "size" );
# kişi sayısının hesap ve bahşişe olan dağılımı
```

```python
sns.scatterplot(data = df , x = "total_bill" ,y = "tip" ,  hue = "size"  , size = "size");
```

```python
sns.scatterplot(data = df , x = "total_bill" , y = "tip" , hue = "day" ,size = "size");
# günler ve kişi sayısına göre bahiş ve hesap ücretlerininn aralarındaki ilişkler ile birlikte dağılımı
```
