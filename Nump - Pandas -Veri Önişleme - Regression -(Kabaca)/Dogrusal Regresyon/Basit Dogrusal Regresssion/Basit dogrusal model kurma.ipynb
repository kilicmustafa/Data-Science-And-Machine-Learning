```python
import pandas as pd 
df = pd.read_csv("original.csv")
df = df.iloc[: , 1 :len(df)]
df.head()
```

```python
df.info() # dataframe hakkında bilgi verir
```

```python
import seaborn as sns

sns.jointplot(x = "TV" ,y = "sales" , data = df , kind="reg");
# x = bagimsiz degisken , y = bagimli degisken  , data = Dataframe , kind = modelleme türü
```

```python
sns.jointplot(x = "TV" ,y = "sales" , data = df , kind="reg");
```

```python
from sklearn.linear_model import LinearRegression
```

```python
x = df[["TV"]]
```

```python
x.head()
```

```python
y = df[["sales"]]
```

```python
reg = LinearRegression() # nesne oluşturuyoruz
```

```python
model = reg.fit(x , y) # modeli x ve y ile fit ediyoruz
```

```python
model
```

```python
str(model)
```

```python
dir(model)
```

```python
model.intercept_ #b0 katsayısı 
```

```python
model.coef_ # b1 katsayısı
```

```python
# rkare = bagımsız değişkenin bağımlı değişkendeki etki yüzdesidir
model.score(x ,y) 
```
