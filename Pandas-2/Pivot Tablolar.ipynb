```python
import seaborn as sns 
import pandas as pd 
import numpy as np

df = sns.load_dataset("titanic")
df.head()
```

```python
df.groupby("sex")["sex"].count()
# cinsiyete göre yaşayanların sayısı
```

```python
df.groupby(["sex" ,"class"])["survived"].count()
# cinsiyet ve sınıflara göre yaşayanların sayısı
```

```python
df.groupby(["sex" ,"class"])[["survived"]].count() 
#eger dataframe içerisinde görmek istersek "survived" 'i iki köşeli parantez arasına almamız gerekiyor
```

```python
df.groupby(["sex" , "class" ])[["survived"]].mean()
```

```python
df.groupby(["sex" ,"class"])[["survived"]].aggregate("mean")
```

```python
df.groupby(["sex" ,"class"])[["survived"]].aggregate("mean").unstack()
# üsttekinin farklı bir gösterimi
```

```python
df.groupby(["sex" ,"class"])[["survived"]].aggregate("mean").T
# üstekinin saçma ama farklı bir gösterimi
```

```python
df.pivot_table(values = ["survived"] , index = ["sex"] )
#cinsiyete göre yaşama ortalaması pivot table
```

```python
df.pivot_table(values = ["survived"] , index  = ["sex" ,"class"])
# cinsiyet ve sınıflara göre yaşama ortalamaları 
```

```python
df.pivot_table(values = ["survived"] ,index = ["sex"] , columns = ["class"])
```

```python
age = pd.cut(df.age , [0 ,18,90])
```

```python
age.head()
```

```python
df.pivot_table(values = ["survived"] ,index = ["sex",age] , columns= ["class"])
# cınsıyet yaş ve sınıfa göre yaşama oranları
```

```python
fare = pd.qcut(df["fare"] , 2)
fare.head(5)
```

```python
df.pivot_table(values = "survived"  , index = ["sex" ,age] , columns = [fare,"class" ])
```

```python
df.pivot_table(values = "survived" , index = ["sex" , age] ,columns = [fare , "class"] , aggfunc = sum )
#heasplama fonksiyonunun türünu mean dan sum a dönderdim 
```

```python
df.pivot_table(index = ["sex","class"] 
               ,aggfunc = {
                    
                   "survived" : sum,
                   "fare" : "mean"
    
                }
)
# burada is hedef değiskenleri aggfunc içerisinde hesaplamaları ile birlikte verdim
```

```python
df.pivot_table(index = ["sex" ,"class"] ,
              aggfunc = {
                      
                  "survived" : "count",
                  "alive" : "count"
                  
              },margins = True
        )
```
